# ЛАБОРАТОРНАЯ РАБОТА №4

## ИНТЕРПОЛЯЦИЯ И АППРОКСИМАЦИЯ ФУНКЦИЙ

```python
import math

rr = 10000
x0, xn = 0.0, 1.0
n = 20
x = [x0, (x0 + xn) / 2.0, xn]
y = [0] * 3
hx = (xn - x0) / n
ht = 2.0 / n
a = [0] * 3


def calculate_function_value(x_value):
    return round((2.4 * math.cos(x_value) * math.exp(-x_value / 3)) * 1000) / 1000
```

```python
def print_interpolation_points():
    y[0] = calculate_function_value(x[0])
    y[1] = calculate_function_value(x[1])
    y[2] = calculate_function_value(x[2])
    print("-------------------------------------------------")
    print("| xi |" + " x0 = " + f"{x[0]:^4}" + "| x1 = " + f"{x[1]:^4}" + "| x2 = " + f"{x[2]:^4}|")
    print("-------------------------------------------------")
    print("| yi |" + f"{y[0]:>10.3f}" + "|" + f"{y[1]:>10.3f}" + "|" + f"{y[2]:>10.3f}" + "|")
```

```python
def lagrange_interpolation(_x):
    a[0] = y[0] / ((x[0] - x[1]) * (x[0] - x[2]))
    a[1] = y[1] / ((x[1] - x[0]) * (x[1] - x[2]))
    a[2] = y[2] / ((x[2] - x[0]) * (x[2] - x[1]))
    result = a[0] * (_x - x[1]) * (_x - x[2]) + a[1] * (_x - x[0]) * (_x - x[2]) + a[2] * (_x - x[0]) * (_x - x[1])
    return round(result * rr) / rr
```

```python
def first_newton_interpolation(t):
    dy = [y[1] - y[0], y[2] - y[1]]
    result = y[0] + t * dy[0] + t * (t - 1) / 2 * (dy[1] - dy[0])
    return round(result * rr) / rr


def second_newton_interpolation(t):
    dy = [y[1] - y[0], y[2] - y[1]]
    result = y[2] + t * dy[1] + t * (t + 1) / 2 * (dy[1] - dy[0])
    return round(result * rr) / rr
```

```python
def quadratic_interpolation_coefficients():
    a0 = [0] * 5
    det = [0] * 4

    a0[1] = x[0] + x[1] + x[2]
    a0[2] = (x[0] * x[0]) + (x[1] * x[1]) + (x[2] * x[2])
    a0[3] = (x[0] * x[0] * x[0]) + (x[1] * x[1] * x[1]) + (x[2] * x[2] * x[2])
    a0[4] = (x[0] * x[0] * x[0] * x[0]) + (x[1] * x[1] * x[1] * x[1]) + (x[2] * x[2] * x[2] * x[2])

    sumy = y[0] + y[1] + y[2]
    a1 = x[0] * y[0] + x[1] * y[1] + x[2] * y[2]
    a2 = x[0] * x[0] * y[0] + x[1] * x[1] * y[1] + x[2] * x[2] * y[2]

    det[0] = 3 * a0[2] * a0[4] + a0[1] * a0[3] * a0[2] + a0[2] * a0[1] * a0[3] - a0[2] * a0[2] * a0[2] - a0[3] * a0[3] * 3 - a0[4] * a0[1] * a0[1]
    det[1] = sumy * a0[2] * a0[4] + a0[1] * a0[3] * a2 + a0[2] * a1 * a0[3] - a2 * a0[2] * a0[2] - a0[3] * a0[3] * sumy - a0[4] * a1 * a0[1]
    det[2] = 3 * a1 * a0[4] + sumy * a0[3] * a0[2] + a0[2] * a0[1] * a2 - a0[2] * a1 * a0[2] - a2 * a0[3] * 3 - a0[4] * a0[1] * sumy
    det[3] = 3 * a0[2] * a2 + a0[1] * a1 * a0[2] + sumy * a0[1] * a0[3] - a0[2] * a0[2] * sumy - a0[3] * a1 * 3 - a2 * a0[1] * a0[1]

    a[0] = det[1] / det[0]
    a[1] = det[2] / det[0]
    a[2] = det[3] / det[0]
```

```python
def quadratic_approximation_polynomial(_x):
    quadratic_interpolation_coefficients()
    result = a[2] * _x * _x + a[1] * _x + a[0]
    return round(result * rr) / rr
```

```python
def tabulate_function_values():
    _x = x[0]
    t1 = 0.0
    t2 = -2.0

    print("---------------------------------------------------------------------------------------------------------")
    print("| i |   xi   |   ti   |   t'i    |   f(xi)   |   L2(xi)  ||f(xi) - L2(xi)||   N1(xi)   ||f(xi) - N1(xi)||")
    print("---------------------------------------------------------------------------------------------------------")

    for i in range(n + 1):
        _x = round(_x * rr) / rr
        t1 = round(t1 * rr) / rr
        t2 = round(t2 * rr) / rr
        print(
            f"|{i:^3}|{_x:^8.4f}|{t1:^8.4f}|{t2:^10.4f}|{calculate_function_value(_x):^11.4f}|{lagrange_interpolation(_x):^11.4f}|{round(abs(calculate_function_value(_x) - lagrange_interpolation(_x)) * rr) / rr:^16.4f}|{first_newton_interpolation(t1):^12.4f}|{round(abs(calculate_function_value(_x) - first_newton_interpolation(t1)) * rr) / rr:^16.4f}|")
        _x += hx
        t1 += ht
        t2 += ht

    print("---------------------------------------------------------------------------------------------------------")
    print("| i |   xi   |   ti   |   t'i    |   f(xi)   |   N2(xi)  ||f(xi) - N2(xi)||   P2(xi)   ||f(xi) - P2(xi)||")
    print("---------------------------------------------------------------------------------------------------------")

    _x = x[0]
    t1 = 0.0
    t2 = -2.0

    for i in range(n + 1):
        _x = round(_x * rr) / rr
        t1 = round(t1 * rr) / rr
        t2 = round(t2 * rr) / rr
        print(
            f"|{i:^3}|{_x:^8.4f}|{t1:^8.4f}|{t2:^10.4f}|{calculate_function_value(_x):^11.4f}|{second_newton_interpolation(t2):^11.4f}|{round(abs(calculate_function_value(_x) - second_newton_interpolation(t2)) * rr) / rr:^16.4f}|{quadratic_approximation_polynomial(_x):^12.4f}|{round(abs(calculate_function_value(_x) - quadratic_approximation_polynomial(_x)) * rr) / rr:^16.4f}|")
        _x += hx
        t1 += ht
        t2 += ht


print_interpolation_points()
tabulate_function_values()
```
