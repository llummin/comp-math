### Система линейных алгебраических уравнений
$$\begin{cases} 
3x - y - z = -1 \\ 
2x - 4y + z = -4 \\
2x + 3y - 6z = 3
\end{cases}$$

### Программная реализация методов

```python
import math

eps = 0.001  # точность
x, y, z = -1 / 3, 4 / 4, -3 / 6  # начальное приближение
```

### Метод простых итераций

```python
def method_simple_implications():
    k = 0
    x0, x1 = x, x  # x0 = x(k), x1 = x(k+1)
    y0, y1 = y, y  # y0 = y(k), y1 = y(k+1)
    z0, z1 = z, z  # z0 = y(k), z1 = z(k+1)
    modx, mody, modz = 0, 0, 0  # modx: |x1 - x0|, mody: |y1 - y0|, modz: |z1 - z0|
    print("| k |  x^(k)  | x^(k+1) ||x^(k+1) - x^(k)||  y^(k)  | y^(k+1) ||y^(k+1) - y^(k)||  z^(k)  | z^(k+1) ||z^(k+1) - z^(k)||")
    print("|---|---------|---------|-----------------|---------|---------|-----------------|---------|---------|-----------------|")
    while True:
        x0, y0, z0 = x1, y1, z1
        x1 = (y0 + z0 - 1) / 3
        y1 = (2 * x0 + z0 + 4) / 4
        z1 = (2 * x0 + 3 * y0 - 3) / 6
        modx = abs(x1 - x0)
        mody = abs(y1 - y0)
        modz = abs(z1 - z0)
        print(f"|{k:2d} |{x0:8.4f} |{x1:8.4f} |{modx:16.4f} |{y0:8.4f} |{y1:8.4f} |{mody:16.4f} |{z0:8.4f} |{z1:8.4f} |{modz:16.4f} |")
        k += 1
        if modx <= eps and mody <= eps and modz <= eps:
            break
```

```python
method_simple_implications()
```

```python
def method_Seidel():
    k = 0
    x0, x1 = x, x  # x0 = x(k), x1 = x(k+1)
    y0, y1 = y, y  # y0 = y(k), y1 = y(k+1)
    z0, z1 = z, z  # z0 = y(k), z1 = z(k+1)
    modx, mody, modz = 0, 0, 0  # modx: |x1 - x0|, mody: |y1 - y0|, modz: |z1 - z0|
    print("| k |  x^(k)  | x^(k+1) ||x^(k+1) - x^(k)||  y^(k)  | y^(k+1) ||y^(k+1) - y^(k)||  z^(k)  | z^(k+1) ||z^(k+1) - z^(k)||")
    print("|---|---------|---------|-----------------|---------|---------|-----------------|---------|---------|-----------------|")
    while True:
        x0, y0, z0 = x1, y1, z1
        x1 = (y0 + z0 - 1) / 3
        y1 = (2 * x1 + z0 + 4) / 4
        z1 = (2 * x1 + 3 * y1 - 3) / 6
        modx = abs(x1 - x0)
        mody = abs(y1 - y0)
        modz = abs(z1 - z0)
        print(f"|{k:2d} |{x0:8.4f} |{x1:8.4f} |{modx:16.4f} |{y0:8.4f} |{y1:8.4f} |{mody:16.4f} |{z0:8.4f} |{z1:8.4f} |{modz:16.4f} |")
        k += 1
        if modx <= eps and mody <= eps and modz <= eps:
            break
```
